```python
data = sci.loadmat('/content/drive/MyDrive/Urban.mat')
X = data['X']
X = X.astype(float) # Please take care that the data matrix must be cast to float in Python, for subsequential operations

gt = sci.loadmat('/content/drive/MyDrive/end6_groundTruth.mat')
abundances = gt['A']

nCol = 307
nRow = 307

plt.figure()
#TO DO : try to interpret the abundances that you see. Hint : the 6 sources are usually called ['Grass','Road','Rooftop (2)','Dirt','Trees','Rooftop (1)']. Associate a name to each endmember.
names = ['Road','Dirt','Grass','Rooftop (1)','Rooftop (2)','Trees']
for ii in range(6): 
    ax = plt.subplot(2, 3, ii+1)
    ax.imshow(abundances[ii,:].reshape(nCol,nRow))
    
    plt.title(names[ii])
```

```python
endmembers= np.zeros((162,6))

for ii in range(6): 
    ind = np.where(abundances[ii] > 0.999)[0]
    allPP = X[:,ind]
    endmembers[:,ii] = np.mean(allPP,axis=1)


plt.figure(),plt.plot(endmembers)
```

```python
X2 = np.zeros((nCol+1, nRow+1, 162))
X2[:-1, :-1] = X.T.reshape((nCol, nRow, -1))
X2[-1, :], X2[:, -1] = 0, 0
X2.shape
```

```python
n, m = nCol+1, nRow+1
r = 6
nIter = 3000
learning_rate = 0.001

E = endmembers.T
Y = X2.reshape(n*m, 162)
```

```python
DefaultConv2D = partial(tf.keras.layers.Conv2D, kernel_size=3, padding="same",
                        activation=tf.keras.layers.LeakyReLU(alpha=0.1), 
                        kernel_initializer="he_normal")


def build_model(n, m, r):
    inputs = tf.keras.Input(shape=[n,m,r])
    
    skip = DefaultConv2D(filters=4, kernel_size=1)(inputs)
    skip = tf.keras.layers.BatchNormalization()(skip)
    
    main = DefaultConv2D(filters=256, kernel_size=3, strides=2)(inputs)
    main = tf.keras.layers.BatchNormalization()(main)
    main = DefaultConv2D(filters=256, kernel_size=3)(main)
    main = tf.keras.layers.BatchNormalization()(main)
    main = tf.keras.layers.UpSampling2D(size=(2,2), interpolation='bilinear')(main)
    
    main = tf.keras.layers.concatenate([main, skip])
    main = tf.keras.layers.BatchNormalization()(main)
    main =DefaultConv2D(filters=256, kernel_size=3)(main)
    main = tf.keras.layers.BatchNormalization()(main)
    main = DefaultConv2D(filters=256, kernel_size=1)(main)
    main = tf.keras.layers.BatchNormalization()(main)
    
    outputs = DefaultConv2D(filters=r, kernel_size=1, activation='softmax')(main)
    
    return tf.keras.Model(inputs, outputs)


model = build_model(n, m, r)
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
```

```python
def loss_function(A, E, Y):
    return 0.5*tf.reduce_sum((Y - tf.reshape(A, [Y.shape[0], -1]) @ E)**2)
```

```python
A = tf.random.normal(shape=[1,n, m, r])

for i in range(nIter):
    with tf.GradientTape() as tape:
        A = model(A)
        loss = loss_function(A, E, Y)
        
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    print(i)
    
A = model(A)
```
